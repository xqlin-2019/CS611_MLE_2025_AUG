```python
file_name = "features_financials"
df = pd.read_csv(f"data/{file_name}.csv")

# 1Ô∏è‚É£ Count how many snapshot dates each customer appears in
snapshot_counts = df.groupby("Customer_ID")["snapshot_date"].nunique()

# 2Ô∏è‚É£ Find the customer with the highest number of unique snapshot dates
top_customer = snapshot_counts.idxmax()
max_snapshots = snapshot_counts.max()

print(f"{file_name}:\nCustomer with most snapshots: {top_customer} ({max_snapshots} snapshots)")

# 3Ô∏è‚É£ Extract and sort that customer's full history
sub = df[df["Customer_ID"] == top_customer].sort_values("snapshot_date")

# 4Ô∏è‚É£ Display summary and sample
print(f"\nAll snapshot dates for {top_customer}: {sub['snapshot_date'].unique()}")
print(f"Total records: {len(sub)}")

display(sub.head(10))
```
